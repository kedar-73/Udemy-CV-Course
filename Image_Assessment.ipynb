```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread("/content/WIN_20250903_10_19_05_Pro.jpg")
```

```python
plt.imshow(img)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
new_img=img_rgb.copy()
```

```python
#flip the image
new_img=cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
pt1= (800,250)
pt2= (1600,1200)
```

```python
#draw red rectangle around the face
cv2.rectangle(img_rgb,pt1=pt1,pt2=pt2,color=(255,0,0),thickness=10)
plt.imshow(img_rgb)
```

```python
vertices=np.array([[900,1100],[1250,380],[1450,1100]],np.int32)
```

```python
vertices.shape
```

```python
pts=vertices.reshape((-1,1,2))
```

```python
vertices
```

```python
pts.shape
```

```python
#draw triangle
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.polylines(img_rgb,[pts],isClosed=True,color=(0,0,255),thickness=20)
plt.imshow(img_rgb)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.fillPoly(img_rgb,[pts],color=(0,0,255))
plt.imshow(img_rgb)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('/content/WIN_20250903_10_19_05_Pro.jpg')

# Draw a circle at a fixed location (e.g., center)
x, y = img.shape[1]//2, img.shape[0]//2
cv2.circle(img, (x, y), 100, (0, 0, 255), thickness=10)

# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
