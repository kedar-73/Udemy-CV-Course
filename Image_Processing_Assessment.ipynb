```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img=cv2.imread('/content/9708.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
display_img(img)
```

```python
#Apply Binary threshold
img=cv2.imread('/content/9708.jpg')
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
display_img(thresh1,cmap='gray')
```

```python
img=cv2.imread('/content/9708.jpg')
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
display_img(img)
```

```python
#low pass filter with a 4 by 4 kernel
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
```

```python
kernel
```

```python
img=cv2.imread('/content/9708.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
result=cv2.filter2D(img,-1,kernel)
display_img(result)
```

```python
#Horizontal Sobel Filter
img=cv2.imread('/content/9708.jpg',0)
```

```python
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx,cmap='gray')
```

```python
img=cv2.imread('/content/9708.jpg')
color=['b','g','r']
for i, col in enumerate(color):
    histr=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
plt.title('MY HISTO')
```
